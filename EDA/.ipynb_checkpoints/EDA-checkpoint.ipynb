```python
import pandas as pd
import numpy as np
import textblob as tb
import wordcloud
```

```python
data_path = "E:/Workspace/Machine Learning ~ Deep Learning/xdi-amazon-eda-ml/Data"
df = pd.read_csv(data_path + "/Reviews.csv")
df = df.set_index("Id")
```

```python
columns = list(df.columns)
# print(columns)
users = list(df["UserId"].unique()) # 25k+ users
df.head()
```

```python
# taking a lot of time for the dataset.

from nltk.sentiment.vader import SentimentIntensityAnalyzer
users_sentiment = {}

def calc_sentiment():
    user_text_df = df[["UserId", "Summary", "ProfileName"]]
    for index, row in user_text_df.iterrows():
#         print(index)
        userID, review, name = list(row)
#         print(review)
        score = SentimentIntensityAnalyzer().polarity_scores(review)
#         print(score)
        if userID in users_sentiment.keys():
            users_sentiment[userID]['negative'] += score['neg']
            users_sentiment[userID]['neutral'] += score['neu']
            users_sentiment[userID]['positive'] += score['pos']
            users_sentiment[userID]['times'] += 1
        else:
#             print(score['neg'])
            users_sentiment[userID] = {
                'negative' : score['neg'],
                'neutral' : score['neu'],
                'positive' : score['pos'],
                'times' : 1,
                "name" : name
            }

#         print(users_sentiment.keys())
#         break
    print(users_sentiment)
        
calc_sentiment()
```
